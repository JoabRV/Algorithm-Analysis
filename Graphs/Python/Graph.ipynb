```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, vertex1, vertex2, distance):
        if vertex1 not in self.graph:
            self.graph[vertex1] = []
        if vertex2 not in self.graph:
            self.graph[vertex2] = []

        self.graph[vertex1].append((vertex2, distance))
        self.graph[vertex2].append((vertex1, distance))

    def get_neighbors(self, vertex):
        return self.graph.get(vertex, [])

# Create a graph
city_graph = Graph()

# Add edges (city connections) with distances
city_graph.add_edge("A", "B", 10)
city_graph.add_edge("A", "C", 15)
city_graph.add_edge("B", "C", 20)
city_graph.add_edge("B", "D", 25)
city_graph.add_edge("C", "D", 30)

# Get neighbors of a vertex
neighbors_of_B = city_graph.get_neighbors("B")
print("Neighbors of vertex B:", neighbors_of_B)

# Output:
# Neighbors of vertex B: [('A', 10), ('C', 20), ('D', 25)]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo no dirigido
G = nx.Graph()

# Agregar nodos
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Agregar aristas
G.add_edge("A", "B", weight=5)
G.add_edge("B", "C", weight=3)
G.add_edge("C", "D", weight=7)
G.add_edge("D", "A", weight=2)

# Obtener posiciones para visualizaci√≥n
pos = nx.spring_layout(G)

# Obtener etiquetas de pesos de aristas
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}

# Dibujar el grafo
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Mostrar el grafo
plt.title("Ejemplo de Grafo")
plt.axis('off')
plt.show()
```
