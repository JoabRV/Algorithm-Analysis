```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2

# Create sets
set_A = {1, 2, 3, 4, 5}
set_B = {4, 5, 6, 7, 8}

# Create a Venn diagram
venn2(subsets=(len(set_A - set_B), len(set_B - set_A), len(set_A & set_B)),
      set_labels=('Set A', 'Set B'))

# Display the diagram
plt.title("Venn Diagram")
plt.show()
```
